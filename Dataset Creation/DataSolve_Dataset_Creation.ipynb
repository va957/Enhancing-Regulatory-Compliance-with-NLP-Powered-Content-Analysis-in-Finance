```python
df=pd.read_csv("/content/train.csv")
```

```python
df
```

```python
df_new=df.drop_duplicates(subset = "document_text") #Dropping the duplicate rows 
#Therefore the rows come down from 47102 to 9151 Rows
```

```python
df["cat_name"].unique() #Listing the unique categories
```

```python
df_new[df["cat_name"].unique()]=0 #Creating new columns for each of the 50 categories and initializing them to 0
```

```python
df_new
```

```python
for i,row_new in df_new.iterrows():
  df_temp=df[df["document_text"]==row_new["document_text"]]
  for label in df_temp["cat_name"]:
    df_new[label].loc[i]=1
#Assigining 1 as a label to the categories which were orignally labeled for that paticular "document_text"
```
